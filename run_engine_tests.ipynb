# Run Engine Tests

This notebook runs the engine unit tests using subprocess and prints the results.

```python
import subprocess
from pathlib import Path

root = Path('.')
python = root / '.venv' / 'Scripts' / 'python.exe'
commands = [
    [str(python), 'engine/tests/test_gradients.py'],
    [str(python), 'engine/tests/test_attention_gradients.py'],
    [str(python), 'eval/baseline/test_regression.py'],
]

for command in commands:
    print('=== Running:', ' '.join(command), '===')
    result = subprocess.run(command, capture_output=True, text=True)
    print('RETURN CODE:', result.returncode)
    print('STDOUT:')
    print(result.stdout)
    print('STDERR:')
    print(result.stderr)
    print('=== End ===\n')
```
